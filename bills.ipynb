```python
bills=seaborn.load_dataset("tips")
pyplot.scatter(bills.total_bill, bills.tip)
# as Seaborn is closely integrated with pandas data structures, its own scatter plot function unlocks additional features
seaborn.scatterplot(x="total_bill", y="tip", data=bills)
seaborn.scatterplot(x="total_bill", y="tip", hue="smoker", data=bills)
seaborn.scatterplot(x="total_bill", y="tip", hue="size", style="smoker", data=bills)
```
